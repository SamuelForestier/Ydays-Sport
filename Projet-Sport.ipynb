```python
for i in range(len(df)):
    if df['MATCHUP'][i][4:5] == '@':
        df['HOME'][i] = 1
    else:
        df['AWAY'][i] = 1
```

Changement côte US -> EUROPE

```python
def CHANGE_ODDS(x):
    return round(1 + x / 100 , 2) if x > 0 else round(1 + 100 / np.abs(x),2)
```

```python
df_merge['New_ML'] = df_merge['ML'].apply(lambda x : CHANGE_ODDS(x))
df_merge
```
