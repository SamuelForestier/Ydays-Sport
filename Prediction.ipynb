```python
# Importation des bibliothèques
import numpy as np
import pandas as pd
```

```python
# Récupération du jeu de données
df = pd.read_csv('preprocessed_data.csv')
df.head()
```

```python
# Sélection des caractéristiques et de la cible
X = df[["HOME", "AWAY", "FG_PCT_HOME", "FG_PCT_AWAY", "FG3_PCT_HOME", "FG3_PCT_AWAY", "FT_PCT_HOME", "FT_PCT_AWAY", "PLUS_MINUS_HOME", "PLUS_MINUS_AWAY", "EFF_HOME", "EFF_AWAY", "ODDS_HOME", "ODDS_AWAY"]].copy()
y = df["WIN"]

# Encodage des variables catégorielless
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
X[X.columns[X.columns.get_loc('HOME')]] = labelencoder.fit_transform(X['HOME'])
X[X.columns[X.columns.get_loc('AWAY')]] = labelencoder.fit_transform(X['AWAY'])
X.head()
```

```python
# Split des données en train et test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Normalisation des données
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[:5]
```

```python
# Création du modèle et entraînement
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
clf.fit(X_train, y_train)

# Prédiction des résultats
y_pred = clf.predict(X_test)

# Matrice de confusion et score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Matrice de confusion : [[', cm[0][0], ',', cm[0][1], '] [', cm[1][0], ',', cm[1][1], ']]')
print('Score :', clf.score(X_test, y_test))
```

```python
# Création du modèle et entraînement
from xgboost import XGBClassifier
bst = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic')
bst.fit(X_train, y_train)

# Prédiction des résultats
y_pred = bst.predict(X_test)

# Matrice de confusion et score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Matrice de confusion : [[', cm[0][0], ',', cm[0][1], '] [', cm[1][0], ',', cm[1][1], ']]')
print('Score :', bst.score(X_test, y_test))
```
